```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
hd=pd.read_csv('heart-disease.csv')
hd
```

```python
hd.head(10)
```

```python
plt.hist(hd['age'],bins=50);
```

```python
hd.plot.hist(subplots=True,figsize=(10,10));
```

```python
over=hd[hd['age']>50]
over
```

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.scatter(over['age'],over['chol'],over['target'])
ax.set(title='heart disease',xlabel='age',ylabel='chol');
ax.axhline(linestyle="--")
```

```python
plt.style.available
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(over["age"],
                     over["chol"],
                     c=over["target"],
                     cmap="autumn")
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target")
ax.axhline(over["chol"].mean(),
           linestyle="--",
           color="red");
```
